```python
# %matplotlib notebook
%matplotlib inline
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = xr.open_dataset('/Users/tantanmeow/Desktop/WORK/2018-2018/larry-data/sp_model_bottom_velocities.nc')
data
```

```python
from matplotlib import cm
cmap1=cm.Reds
cmap1.set_bad('gray',1.)

spd = np.sqrt(data.v.data**2 + data.u.data**2)
spd[data.BottomDepth.data<4400] = np.nan

fig, ax = plt.subplots(figsize=(10,10))
cb = ax.pcolormesh(-data.lon.data, -data.lat.data, spd, cmap=cmap1, vmin=0, vmax=0.2)
clb = fig.colorbar(cb, ax=ax)
clb.set_label('Speed (ms$^{-1}$)',fontsize=20)
clb.ax.tick_params(labelsize=12)
plt.gca().set_aspect('equal', adjustable='box')
ax.set_xlabel('Longitude ($^{\circ}$W)', fontsize=20)
ax.set_ylabel('Latitude ($^{\circ}$S)', fontsize=20)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.invert_yaxis()
ax.invert_xaxis()
plt.savefig('Fig1.png', dpi=400)
```
