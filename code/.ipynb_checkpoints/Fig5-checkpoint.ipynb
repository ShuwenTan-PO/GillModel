# Model Topography (1-channel)

The numerical domain size is 60 units in both the x-direction and y-direction, with uniform grid size: $\Delta x=\Delta y=0.1$. The time step is $\Delta t=0.01$. A parabolic channel is introduced near the western boundary of the domain, with a Gaussian
topographic feature varying in the along channel direction:
\begin{equation}
d(x)=h_0+\frac{(x-x_c)^2}{r}e^{-\frac{(y-y_c)^2}{2\sigma^2}}
\end{equation}
, where the ($x_c$,$y_c$) denotes the location of the bottom of the parabola at the sill crest, the bottom height $h_0$ is 0.6, and the length of the channel is denoted by $\sigma$. The boundaries of the parabolic channel gradually join the wall of the domain to its west and a tall plateau to its east. A logistic function makes up the east boundary of the plateau and descends to zero smoothly.

Here we plot an example with $r=(w_s/2)^2/d_s$, where $w_s$=50 km.

In the Samoan Passage case, one spacial and temporal unit roughly represent 30 km and 0.6 day, respectively. 

```python
# Compute scales for nondimensional variables
r = 6.371e+6; unit = 2*np.pi*r/360; Omega=7.292e-5;
g = 4e-4; Di = 1000; lat = -8; q = -1; 
f0 = abs(2*Omega*np.sin(lat/180*np.pi));
Ld = np.sqrt(g*Di)/f0;
D = -q*Di;

# # Best guesses of Samoan Passage  
# wsd = [50e+3,20e+3]; hsd = 600; dsd = 500; 

# # nondimensionalize
# wsn = wsd/Ld; hsn = hsd/Di; dsn = dsd/Di;
# alphs = dsd/np.power((np.array(wsd)/2.0),2)
# rs = f0**2/g/alphs
```

NOTICE: In Karl's model, y is right-left direction, x is down-up direction

But in model visulization, we set x : left-right direcrtion and y : down-up direction

```python
# # Wide channel for example:
# wc = wsn[Topo_WN]; bc = hsn; dc = dsn; rc = 1/rs[Topo_WN];
dc = 0.5
rc = dc/np.power((wc/2.0),2)

# Model domain
# dx = 0.1; dy = 0.1; LX = 60; LY = 60;
x = np.arange(0, LX + 1e-6, dx) # from 0 to 60
y = np.arange(0, LY + 1e-6, dy) 

# Specify geometry parameters:
  # How wide is your island?
Ly = 10

# In y-direction:
  # Where is the center of the channel?
y0 = y[-1] - wc - wc/2
  # Where is the joint of the channel and the plateau?
yc = y[-1] - 2*wc - wc
  # How wide is the right slope of the plateau?
theta_i = 5

# In x-direction:
  # Where is the center of the bump?
xc = 35
  # How wide is the bump?
theta_b = 5
```

```python
# The parabolic channel:
b1 = np.zeros((len(x),len(y)))
# The plateau:
b2 = np.zeros((len(x),len(y)))
# The total topography:
b = np.zeros((len(x),len(y)))
# The initial layer thickness:
h = np.zeros((len(x),len(y)))
h_plt = np.zeros((len(x),len(y)))

for j in range(len(y)):
    for i in range(len(x)):
        if y[j]>=yc:
            b1[i,j] = (bc + rc*(y[j] - y0)**2)*np.exp(-(x[i] - xc)**2/2/theta_b**2)

for j in range(len(y)):
    for i in range(len(x)):
        if y[j]<yc:            
            b2[i,j] = b1[i,-1]*(1 + np.tanh((y[j] - (yc - Ly))/theta_i))/2
    
for j in range(len(y)):
    for i in range(len(x)):
        if y[j]>=yc:
            b[i,j] = b1[i,j]
        else:         
            b[i,j] = b2[i,j]
        if b[i,j]<1:
            h[i,j] = 1.0 - b[i,j]
            h_plt[i,j] = 1.0 - b[i,j]
        else:
            h[i,j] = 1e-10
            
```

```python
from matplotlib import cm
cmap=cm.bone
cmap.set_bad('gray',1.)
levels = np.arange(0, 5+1e-4, .1, dtype=float)

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12, 8), sharey=False)

ax[0].plot(y,b[350,])
ax[0].set_ylim([0,5])
ax[1].plot(y,b[450,])
ax[1].set_ylim([0,5])
ax[2].plot(x,b[:,520])
ax[2].set_ylim([0,1])
ax[-1].contourf(y,x,b,levels,
            cmap=cmap,add_colorbar=True)
```

```python
# build Dataset

Topo = xr.Dataset({'Ly': (Ly),
                   'xc': (xc),
                   'theta_i': (theta_i),
                   'theta_b': (theta_b),
                     'y0': (y0),
                     'yc': (yc)},
                 attrs={'title':'Essentials for model topography, theta are for Gaussian',
                       'description':'Ly: island width; xc: sill-crest; theta_i: plateau slope; theta_b: channel length'})

Topo.to_netcdf(Topo_filepath + Topo_filename + '.nc')
```

```python
# visulize topography at the top of the bump x=xc
  # in model coordinate
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), sharey=True)
ax[0].plot(y, np.squeeze(b1[x==xc,:]), label='b: channel')
ax[0].plot(y, np.squeeze(b2[x==xc,:]), label='b: plateau')
ax[0].set_ylabel('z')

ax[1].plot(y, np.squeeze(b[x==xc,:]), label='b: all')
ax[1].plot(y, np.squeeze(b[x==xc,:] + h[x==xc,:]), label='b+h: all')

for axi in ax:
    axi.set_xlabel('y')
    axi.legend()
#     ax.set(xlim=(52,60)) 
```
